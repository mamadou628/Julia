```python
import numpy as np
import matplotlib.pyplot as plt

def generate_points(R1=1, R2=2, num_points=1000):
    points = []
    while len(points) < num_points:
        # Generation des des points 
        x, y = np.random.uniform(-R2, R2, 2)
        r = np.sqrt(x**2 + y**2)
        if R1 <= r <= R2:
            
            # la proportion des points par r
            if np.random.rand() < r / R2:
                points.append((x, y))
    return np.array(points)

# C'est une fonction a parametre unitialiser donc on peut appeler sans passer les parametre
points = generate_points(num_points=10_000)
plt.scatter(points[:, 0], points[:, 1], s=4) # les deux premier composante
plt.gca().set_aspect('equal') # l'equation qui permet de faire le cercle en round
plt.show()

```

```python
def generate_ellipse_points(a=2, b=1, num_points=1000):
    points = []
    while len(points) < num_points:
        x= np.random.uniform(-a, a)
        y = np.random.uniform(-b, b)
        r = np.sqrt((x/a)**2 + (y/b)**2)

        if r <= 1:
            if np.random.rand() < r/a:  # si le nombre fourni au hazarh est inferieure a r
                points.append((x, y))
    return np.array(points)

points = generate_ellipse_points(a = 4, b=3, num_points=10000)
plt.scatter(points[:, 0], points[:, 1], s=1)
plt.gca().set_aspect('equal')
plt.show()
```

```python
def generate_square_points(L=1, num_points=1000):
    points = []
    max_d = np.sqrt((2*L)**2 + (2*L)**2)  # Le point maxmal des l'angle
    while len(points) < num_points:
        x, y = np.random.uniform(-L, L, 2)
        d = np.sqrt((x - L)**2 + (y - L)**2)
        if np.random.rand() < d / max_d:  # si la condition 
            points.append((x, y))
    return np.array(points)

points = generate_square_points(num_points=10_000)
plt.scatter(points[:, 0], points[:, 1], s=1)
plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def beta_accept_reject(n_samples=1000):
    samples = []
    c = 16/6  # La constante C
    while len(samples) < n_samples:
        x = np.random.uniform(0, 1)  # la fonction de gx
        u = np.random.uniform(0, 1)  # generere u comme probabilite
        if u <= 12 * x * (1 - x)**2 / c:  # verification de la condition
            samples.append(x)
    return np.array(samples)

# 
iteraction = 100000
samples = beta_accept_reject(iteraction)

# la representation graphique
plt.hist(samples, bins=50, density=True, alpha=0.6, label='choix')
x_vals = np.linspace(0, 1, 1000)
plt.plot(x_vals, 12 * x_vals * (1 - x_vals)**2, 'r-', label='Je suis content de comprendre ')
plt.legend()
plt.show()
```
